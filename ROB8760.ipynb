```python
# Adapted from https://huggingface.co/meta-llama/Llama-3.2-3B-Instruct
import torch
from transformers import pipeline

import transformers
import torch

model_id = "meta-llama/Meta-Llama-3-8B-Instruct"

pipeline = transformers.pipeline(
    "text-generation",
    model=model_id,
    model_kwargs={"torch_dtype": torch.bfloat16},
    device_map="auto",
)


for i in range(21):
  messages = [
      {"role": "system", "content": "You are a robot arm tasked with trash detection and collection"},
      {"role": "user", "content": test_cases_one[i]}
  ]

# for i in range(21):
#   messages.append({"role": "user", "content": test_cases_one[i]})

  terminators = [
      pipeline.tokenizer.eos_token_id,
      pipeline.tokenizer.convert_tokens_to_ids("<|eot_id|>")
  ]

  outputs = pipeline(
      messages,
      max_new_tokens=256,
      eos_token_id=terminators,
      do_sample=True,
      temperature=0.6,
      top_p=0.9,
  )
  print(outputs[0]["generated_text"][-1])

# # Test one
# for i in range(21):
#   messages = [
#       {"role": "system", "content": "You are a robot arm tasked with trash detection and collection"},
#       {"role": "user", "content": test_cases_one[i]},
#   ]
#   outputs = pipe(
#       messages,
#       max_new_tokens=256,
#   )
#   print(outputs[0]["generated_text"][-1])

# Test two
# Let's hold off on this test for a sec before we decide: "whether there should be a ground truth", "What should be the ground truth", "Should we decide whether to add an "I don't know"
# for i in range(21):
#   messages = [
#       {"role": "system", "content": "You are a robot arm tasked with trash detection and collection"},
#       {"role": "user", "content": test_cases_two[i]},
#   ]
#   outputs = pipe(
#       messages,
#       max_new_tokens=256,
#   )
#   print(outputs[0]["generated_text"][-1])
```
